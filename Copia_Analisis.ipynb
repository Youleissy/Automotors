```python
#Rename the columns to english languaje

df.rename(columns={'IdCliente':'IdCustomer','NombreCliente':'CustomerName','Fecha':'Date','Empleado':'Employee','Referencia':'Reference','Descripcion':'Description','CodigoFamilia':'FamilyCode','Familia':'Family','Cantidad':'Quantity','Ventas': 'Sales','Localidad':'Location','Sede':'Subsidiary','Area':'Area' }, inplace=True)
```

Some features are in the wrong format ex:



*   Date will be date
*   Quantity will be a float
*   Sales will be a float








```python
#Check the possition of days month and year into the column date in a group of rows, in order to change the formart
df1['Date'].iloc[10000:10003]
```

```python
#'Date' configuation: d/m/Y. Change to date format
df1['Date'] = pd.to_datetime(df1['Date'], format='%d/%m/%Y')
```

```python
# In order to analyze trend sales, separate d/m/y in different columns and quarter of year
df1['Day'] = df1['Date'].dt.day
df1['Month'] = df1['Date'].dt.month
df1['Year'] = df1['Date'].dt.year
df1['Quarter'] = df1['Date'].dt.quarter
df1.head()
```

```python
df1["Quantity"] = pd.to_numeric(df["Quantity"], errors='coerce', downcast="float").astype(float)
```

```python
df1["Sales"] = pd.to_numeric(df["Sales"], errors='coerce', downcast="float").astype(float)
```
